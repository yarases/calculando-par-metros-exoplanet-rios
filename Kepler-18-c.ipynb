# **Exoplaneta Kepler-18-c**

```python
from math import sqrt
import numpy as np
```

**Aqui abaixo estão alguns valores que serão utilizados nas contas a seguir**

```python
M_sol = 1.98892e30                         #Em Kg
G = 6.67e-11                               #Costante Gravitacional
R_sol = 6.9634e8                           #Em Km
R_terra= 6300                              #Em Km
R_jup = 69911                              #Em Km
Luminosidade_sol = 3.8e26                  #Em watts
Albedo = 0.367
Const_Boltz = 5.67037442e-8                #Constante de Boltzmann
UA = 1.496e11                              #Unidades Astronômicas em metros
M_star = 0.927                             #Massa da estrela em massas solares
Depth = 0.0022866                          #Profundidade de trânsito já dividida por 100 pois é dada em %
R_star = 0.951815                          #Em massas solares
Periodo_orbital = 7.64159                  #Em dias
Pi = 3.14159265358979
M_terra = 5.972e24                         #Em Kg
T_sol = 5777                               #Temperatura do Sol em Kelvin
T_star = 5359                              #Temperatura da estrela em Kelvin
K = 5.1                                    #Velocidade Radial em m/s
M_jup = 1.898e27                           #Massa de Júpiter em Kg
```

## Através da fórmula abaixo irei calcular o raio do planeta

raio_exo.jpeg


```python
R_abs_star = (R_star*R_sol)
print('Raio absoluto da estrela em metros', R_abs_star)
print('')
Rp = (R_abs_star)*(sqrt(Depth))
print('O raio do exoplaneta em metros é:', Rp, 'm')
print('')





t = Rp/(R_terra*1000)
print('Em raios terrestres:', '{:.3f}'.format(t))
print('')

j = Rp/(R_jup*1000)
print('Em raios de Júpiter:', '{:.3f}'.format(j))






```

# Calculando o Semi Eixo Maior através da fórmula
semi_eixo.jpeg

Onde **G** é a constante gravitacional, **M** é a massa da estrela e **P** é o período orbital em segundos


```python
Periodo = float(Periodo_orbital*86400)
G = float(6.67e-11)
MassaEstrela = float(M_star*M_sol)
SemiEixoMaior = float(((G*(1.9292524000e30)*(Periodo**2))/(4*(Pi)**2))**(1/3))
print('Semi Eixo Maior em metros:', SemiEixoMaior)
print('')
x = SemiEixoMaior/UA
print('Valor do semi eixo maior em UA:', '{:.4f}'.format(x))
```

## Probabilidade de Trânsito
prob_transito (2).jpeg

Onde **Rs** é o raio da estrela, **e** é a excentricidade e **a** é o semi eixo maior. Porém para este exoplaneta não utilizaremos a expressão que está entre parênteses pois a excentricidade do planeta é 0

```python
Prob_t = (R_abs_star)/SemiEixoMaior
print('Probabilidade de Trânsito é igual a:', Prob_t)
```

# Luminosidade da estrela
luminosidade_formula.jpeg

Precisaremos do valor da luminosidade da estrela para os próximos cálculos
Aqui **R** é o raio da estrela, **sigma** é a constante de Boltzmann e **T** é a temperatura efetiva da estrela

```python
luminosidade_star=float(4*Pi*(R_abs_star**2)*Const_Boltz*(T_star**4))
print('Luminosidade da estrela em watts:', luminosidade_star)
print('')
em_solar = (luminosidade_star/Luminosidade_sol)
print('Luminosidade em Lsol:', em_solar)
print('')

site = np.log10(em_solar)
print('Valor da luminosidade da estrela como está nos parâmetros do site:', '{:.7f}'.format(site))

```

## Temperatura de equilíbrio do planeta 
teq2.jpeg

Onde **T** é a temeperatura efetiva da estrela, **A** é o albedo da Terra, **R** é o raio da estrela e **a** é o semi eixo maior

```python
#utilizarei aqui o valor do SMR em UA e em metros
#no site não tem o valor da temperatura de equilíbrio

Teq = T_star*((1-Albedo)**(1/4))*(2*R_abs_star/UA)**(1/2)
print('Temperatura de equilíbrio do planeta', Teq, 'K')
print('')
Teq2 = T_star*((1-Albedo)**(1/4))*(R_abs_star/(2*SemiEixoMaior))**(1/2)
print('Temperatura de equilíbrio utilizando o smr em metros:',Teq2, 'K')
```

#Zona Habitável do planeta

image.png

Iremos utilizar esses padrões da NASA






```python
Z_interna = 0.75*((luminosidade_star/Luminosidade_sol)**1/2)
Z_central = 1.0*((luminosidade_star/Luminosidade_sol)**1/2)
Z_externa = 1.77*((luminosidade_star/Luminosidade_sol)**1/2)
print('Zona Habitável interna', Z_interna, 'UA')
print('')
print('Zona Habitável central', Z_central, 'UA')
print('')
print('Zona Habitável externa', Z_externa, 'UA')
```

# Massa do planeta

```python
Mp = float((MassaEstrela*K)/((2*Pi*SemiEixoMaior)/(Periodo_orbital*86400)))
jup = Mp/M_jup              #Pega a massa total do exoplaneta e divide pela massa de Júpiter para assim obter a massa em Mjup
terra = Mp/M_terra          #O mesmo procedimento que fiz acima, só que dessa vez com a massa da Terra

print('Massa do exoplaneta:', Mp, 'Kg')
print('')
print('Massa em função de Júpiter:', jup)
print('')
print('Massa em função da Terra:', terra)
```

# Densidade do exoplaneta

```python
Densidade = float(((3*Mp)/((4*Pi)*Rp**3))/1000)
print('A densidade do exoplaneta é:', Densidade , 'g/cm³')
```

# Tempo de Trânsito

```python
Temp_t=float(((sqrt(SemiEixoMaior*G*MassaEstrela))*(2*R_abs_star))/(G*MassaEstrela))
print('Tempo de trânsito em segundos:', "%.10e"%float(Temp_t))

horas = Temp_t/3600
dias = Temp_t/86400
print('Tempo de trânsito em horas:', horas)
print('Tempo de trânsito em dias:', dias)
```
